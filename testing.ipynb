```python
import sqlite3
conn = sqlite3.connect('banking_assistant.db')
cursor = conn.cursor()

# Check exact count
cursor.execute("SELECT COUNT(*) FROM products WHERE bank_name='HDFC' AND category='Credit Card'")
print(f"Database has: {cursor.fetchone()[0]} HDFC credit cards")

# List all to see duplicates
cursor.execute("SELECT product_name FROM products WHERE bank_name='HDFC' AND category='Credit Card' ORDER BY product_name")
products = [row[0] for row in cursor.fetchall()]
print(f"Products: {products}")
print(f"Unique names: {len(set(products))}")
```
